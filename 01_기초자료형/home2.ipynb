```python
fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}

# 1. 모든 과일 이름(key)을 리스트로 출력하시오

# 2. 모든 가격(value)을 리스트로 출력하시오

# 3. 'banana'의 가격을 출력하시오

# 4. 'melon'이 딕셔너리에 있는지 확인하시오

# 5. 'grape'를 pop() 함수를 사용하여 삭제하고, 삭제된 값을 출력하시오

fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}
temp = fruits.pop('grape')
print("1)",list(fruits.values()))
print("2)",list(fruits.keys()))
print("3)",fruits.get('banana'))
print("4)","melon" in fruits)
print("5)",temp)
```

```python
# 다음 학생들의 성적 정보를 딕셔너리로 저장하시오
# 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
# 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
# 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

# 요구사항:
# 1. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가
# 2. 각 학생의 정보를 보기 좋게 출력

# *딕셔너리에 추가를 하지않았음.*
student1 = {"name" : "김민수" , "국어": 85 , "영어" : 90 , "수학": 78}
student2 = {"name" : "이영희" , "국어": 92 , "영어" : 88 , "수학": 95}
student3 = {"name" : "박철수" , "국어": 78 , "영어" : 85 , "수학": 82}

총점1 = student1["국어"] + student1["영어"] +student1["수학"]  
총점2 = student2["국어"] + student2["영어"] +student2["수학"]
총점3 = student3["국어"] + student3["영어"] +student3["수학"]
평균1 = 총점1 / 3
평균2 = 총점2 / 3
평균3 = 총점3 / 3

student1['총점'] = 총점1
student2['총점'] = 총점2
student3['총점'] = 총점3

student1['평균'] = round(평균1, 2)
student2['평균'] = round(평균2, 2)
student3['평균'] = round(평균3, 2)

print(f"{student1['name']} - 국어: {student1['국어']}점, 영어: {student1['영어']}점, 수학: {student1['수학']}점 총점: {student1['총점']}, 평균: {student1['평균']}")
print(f"{student2['name']} - 국어: {student2['국어']}점, 영어: {student2['영어']}점, 수학: {student2['수학']}점 총점: {student2['총점']}, 평균: {student2['평균']}")
print(f"{student3['name']} - 국어: {student3['국어']}점, 영어: {student3['영어']}점, 수학: {student3['수학']}점 총점: {student3['총점']}, 평균: {student3['평균']}")
```

```python
text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 각 단어의 등장 횟수 세기
# 3. 결과를 출력


word = text.split()

print(word)
print("python:",word.count('python'))
print("is:",word.count('is'))
print("great",word.count('great'))
print("fun",word.count('fun'))
print("powerful:",word.count('powerful'))
```
