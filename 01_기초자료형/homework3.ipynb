```python
### 문제 2-1: 학점 계산기

# 점수를 입력받아 다음 기준으로 학점을 출력하세요:

# - 90점 이상: A
# - 80점 이상 90점 미만: B
# - 70점 이상 80점 미만: C
# - 60점 이상 70점 미만: D
# - 60점 미만: F
#   score = int(input("점수를 입력하세요: "))
#   **예상 출력:**
#   점수를 입력하세요: 85
#   학점: B

# 여기에 코드를 작성하세요
score = int(input("점수를 입력하세요: "))
if score>=90:
    print("A")
elif score>=80:
    print("B")
elif score>=70:
    print("C")
elif score>=60:
    print("D")
else:
    print("F")
```

```python
## 보너스 문제(선택)

### 윤년 판별기

# 연도를 입력받아 윤년인지 아닌지 판별하세요.

# 윤년 조건:

# 1. 4로 나누어떨어지는 해는 윤년
# 2. 단, 100으로 나누어떨어지는 해는 평년
# 3. 단, 400으로 나누어떨어지는 해는 윤년

# year = int(input("연도를 입력하세요: "))
# **예시:**

# - 2020년: 윤년 (4로 나누어떨어짐)
# - 2100년: 평년 (100으로 나누어떨어지지만 400으로는 안됨)
# - 2000년: 윤년 (400으로 나누어떨어짐)

year = int(input("연도를 입력하세요: "))

if year%400 == 0:
    print("윤년")
else:
    if year%100 == 0:
        print("평년")
    else:
        if year%4 ==0:
            print("윤년")
        else:
            print("평년")
```
