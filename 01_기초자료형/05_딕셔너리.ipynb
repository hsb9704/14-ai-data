```python
# 딕셔너리 쌍 추가  #리스트와 딕셔너리의 요소를 추가하는 방법의 차이 리스트는 인덱스를 사용해 추가 딕셔너리에는 인덱스의 개념이 없음 #리스트는 객체로 들어갈 수 없음 
a = {1: "a"}
a[2]="b"
print(a)
a["name"] = "ken"
print(a)

a[(1,2)] = "hobby"
print(a)


a[[3,4]] = "job"
print(a)
```

```python
#  딕셔너리 요소 삭제하기 key값적기
print(a)
del a[1]
print(a)

del a["name"]
print(a)
```

```python
# 딕셔너리에서 key를 사용해 value 얻기

profile = {
    "name":"홍길동",
    "age": 25 }
print(profile["name"])
```

```python
#딕셔너리와 관련된 함수
# 1.key리스트 만들기 -keys 
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}
print(profile.keys())


# 2. valuse 리스트 만들기 -valuse
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}
print(profile.values())


# 3. key value 쌍 얻기 - items
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}           
print(profile.items())


#4. key 와 value 쌍 모두 지우기 -clear
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}  
print(profile.clear())      


# 5. key로 value 얻기 --get
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}  
print(profile.get("이름"))

# 해당 key가 딕셔너리안에 있는지 조사하기 - in
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}  
print(profile.get("이름"))

print("이름"in profile)


# key로 value 얻기 -pop
profile = {
    "이름":"홍길동",
    "나이": 25 ,"번호":"010 -2123-1234"}  
print(profile.get("이름"))

temp = profile.pop("이름")
print(temp)
print(profile)


  


```
