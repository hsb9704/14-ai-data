```python
scores = [85, 92, 78, 90, 88, 76, 95, 89]
길이 = len(scores)
print(길이)
최댓값 = max(scores)
print(최댓값)
최솟값 = min(scores)
print(최솟값)
총합 = sum(scores)
print(총합)
평균값 = sum(scores) / len(scores)
print(평균값)
정렬 = sorted(scores)
print(정렬)
내림차순정렬 = sorted(scores, reverse=True)
```
