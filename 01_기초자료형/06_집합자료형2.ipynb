```python
# 교집합
s1 =set([1,2,3,4,5,6])
s2 =set([4,5,6,7,8,9])
print(s1.intersection(s2))
print(s1&s2)
```

```python
# 합집합   중복제거

s1 =set([1,2,3,4,5,6])
s2 =set([4,5,6,7,8,9])

print(s1.union(s2))
print(s1|s2)
```

```python
# 차집합 두개의 다른점
s1 =set([1,2,3,4,5,6])
s2 =set([4,5,6,7,8,9])

print(s1.difference(s2))
print(s2.difference(s1))
print((s1 - s2))
print((s2 - s1))
```

```python
#  값 1개 추가 --add
s1 = set([1,2,3])
s1.add(4)
print(s1)
```

```python
#  값을 여러개 추가하기 --update

s1 = set([1,2,3])

s1.update([4,5,6])
print(s1)


```

```python
# 특정값 제거 --remove

s1 = set([1,2,3])

s1.remove(2)
print(s1)
```

```python
#모든값 제거하기
s1 = set([1,2,3])
s1.clear
print(s1)
```
