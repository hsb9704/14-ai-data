```python
#입력값이 몇개가 될지 모를때는 어떻게 할까

def add_many(*args):
    result =0
    print(type(args))
    print(args)
    for i in args:
        result += i
    return result

result = add_many(1,2,3,4,5)

print(result)
print()
```

```python

def add_mul(choice, *args):
   
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":
        result =1
        for i in args:
            result = result*i
    return result

result = add_mul("mul", 1,2,3,4,5)
print(result)
```

```python
# 키워드 매개변수, kwarges 
def print_kwarges(**kwargs):
    print(kwargs)
```

```python
def create_profile(**kwargs):
    print("==== 프로필 정보 ====")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
create_profile(이름="김철수", 취미= "축구",나이=25)
```

```python
# 매개변수에 초기값 미리 설정하기

def say_myself(name, age, man=True):
    print(f"너의 이름은 {name}이다")
    print(f"너의 나이는 {age}이다")
    if man:
        print("남자입니다")
    else:
        print("여자입니다")

say_myself("홍길동" ,25)


say_myself("김영희", "30", False)
```

```python
# lambda 예약어
# 함수_이름 = lambda 매개변수1, 매개변수2, ....: 수행할 문장

def add(a, b):
    return a + b

add = lambda a,b: a+b
```

```python
# 함수의 독스트림(Docstring)  #

def add(a, b):
    """
    두 숫자를 더하는 함수
    Parameters:
    a(int, float) : 첫 번째 숫자
    b(int, float) : 두 번째 숫자

    returns:
    int , float : 두 숫자의 함
    """

    return a+b

print(add.__doc__)
```
