```python
#파이썬함수의 구조

# def 함수이름(매개변수1, 매개변수2, ...):
#     수행할 문장1
#     수행할 문장2
#     ...

def add(a, b):
    return a+b
```

```python
result= add(10 , 20)
print(result)
```

```python
# 4가지 종류의 함수
# 1. 일반적인 함수 : 입력값이 있고, 반환값이 있다.
def add(a,b):
    result = a+b
    return result

sum=add(3,4)
print(sum)
```

```python
#2. 입력값이 없는 함수: 입력값 X, 반환값 O

def say():
    return "Hi"
a =say()
print(a)
```

```python
# 3.반환값이 없는 함수 

def add(a, b):
    print(f"{a},{b}의 합은 {a+b}입니다")

sum = add(3,4)
print(sum)
```

```python
# 4.입력값과 반환값 둘다 없는 함수
def say():
    print("hi")

a =say()
print(a)
```
