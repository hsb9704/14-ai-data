```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result = self.result + num
        return self.result
    def sub(self, num):
        self.result -= num
        return self.result
    def mul(self, num):
        self.result *= num
        return self.result
    def div(self, num):
        self.result /= num
        return self.result
    


cal4 = Calculator()
cal5 = Calculator()
cal6 = Calculator()

cal4.add(10)
cal4.sub(2)
cal4.mul(5)
print(cal4.div(4))

cal5.add(5)
cal5.mul(10)
cal5.sub(10)
print(cal5.div(5))


```
