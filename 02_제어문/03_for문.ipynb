```python
#for문

marks = [90,25,67,45,80]
number= 0
for mark in marks:
    number = number+1
    if mark >= 60:
        print(f"{number}학생은 합격입니다")
    else:
        print(f"{number}학생은 불합격입니다")
```

```python
#for문과 continue문

marks = [90,25,67,45,80]
number= 0


for mark in marks:
    number += 1
    if mark <60:
        continue
    print(f"{number}학생은 합격입니다")
```

```python
marks = [90,25,67,45,80]                                   


for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f"{number+1}학생은 합격입니다")
```

```python
# enumerate()함수 활용하기

# fruits = ["apple", "banana" , "orange"]
# for i in fruits:
#     print(f"번째 : {i}")


fruits = ["apple", "banana" , "orange"]
for i in enumerate [fruits]:
    print(f"{i}번째 :" , fruits)
```

```python
#zip 함수로 여러 리스트 함께 순회하기

names = ["홍길동","유재석","강호동"]
scores = [60 ,70 ,80]
for name , score in zip(names, scores):
    print(f"{name}:{score}점")
```

```python
# 3개 이상의 리스트 --zip()
names = ["홍길동","유재석","강호동"]
korean = [60,70,80]
English =[80,90 ,70]

for name, score1 ,score2 in zip(names,korean,English):
    print(f"{name}: 국어:{score1}점, 영어:{score2}점")
```
