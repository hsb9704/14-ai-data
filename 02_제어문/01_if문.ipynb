```python
#성적이 90점이상이면 "A학점"
#성적이 90점미만이면서 80점 이상이면 "B학점"
#성적이 80점미만이면 70점 이상이면"C학점"
#성적이 70점미만이면 60점 이상이면"D학점"
#그외엔 "F학점"


score = 75

if score >=90:
    print("A학점")
elif score >= 80:
    print("B학점")
elif score >= 70:
    print("C학점")
elif score >= 60:
    print("D학점")
else:
    print("F학점")



```

```python
#match-case 구문

#match 변수
#   case 패턴1:
#       수행할 문장
#   case 패턴2:
#       수행할 문장


grade = input("A ~ F 중 하나의 값을 입력하시오")

match grade:
    case "A"|"B"|"C":
        print("합격입니다")
    case _:
        print("불합격입니다")


```
