```python
# 반복문 ---while
# while의 기본구조

# while 조건문:
#     수행할문장1
#     수행할문장2
#     수행할문장3
#     ...
```

```python
treehit = 0

while treehit < 10:
    treehit = treehit + 1
    print(f"나무를 {treehit}번 찍었습니다.")
    if treehit==10:
        print("나무가 넘어갑니다")
  
```

```python
number =0
prompt = """
1.Add   
2.Del
3.Lsit
4.Quit"""



while number != 4:
    print(prompt)
    number = int(input("1~4까지의 숫자 중 하나를 입력해주세요."))
```

```python
# while문을 강제로 빠져나가기

coffee = 5
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee -= 1
    print(f"남은 커피는 {coffee}개 입니다")
    if coffee ==0:
        print("커피를 다 팔았습니다. 판매를 중지합니다")
        break
```

```python
coffee = 3


while True:
    money =int(input("돈을 넣어주세요"))
    if money ==300:
        print("커피를 줍니다")
        coffee -= 1
    elif money >300:
        print(f"거스름돈을 {money -300}원을 주고, 커피를 줍니다")
        coffee -= 1
    else:
        print("돈을 다시 돌려주기 커피를 주지않는다")
        print(f"현재 남은 커피는 {coffee}개 입니다")

    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중단합니다")
        break

```

```python
#  while문의 맨 처음으로 돌아가기 -continue
# 1~10중에서 홀수만 출력하기

n = 0
while n <10:
    n = n+1 # n+=1
    if n %2 == 0:
        continue
    print(n)
```
