```python
# 반복문 ---while
# while의 기본구조

# while 조건문:
#     수행할문장1
#     수행할문장2
#     수행할문장3
#     ...
```

```python
treehit = 0

while treehit < 10:
    treehit = treehit + 1
    print(f"나무를 {treehit}번 찍었습니다.")
    if treehit==10:
        print("나무가 넘어갑니다")
  
```

```python
number =0
prompt = """
1.Add   
2.Del
3.Lsit
4.Quit"""



while number != 4:
    print(prompt)
    number = int(input("1~4까지의 숫자 중 하나를 입력해주세요."))
```
